#### 7. Handling Missing data

>+ **Droping rows with more than 3 missings value**

```python
new= df[df.isna().sum(axis=1) < 4]
new.isna().sum(axis=1).plot(kind='bar')
```

```python
print("Old data frame length:", len(df))
print("New data frame length:", len(new)) 
print("Number of rows with at least 3 NA value: ", (len(df)-len(new)))
```

*Findings : 35 rows are deleted from the data which stands for 2% of data*

```python
new.isna().sum().plot(kind='bar')
```

>+ **Filling empty cells of no. of reviews column with 0**
>+ **This means that no one has given any review yet**

```python
new["num_of_reviews"].fillna(0, inplace = True) 
new.isna().sum().plot(kind='bar')
```

>+ **Filling empty cells of comments length column with 0** 
>+ **No reviews so no comments**

```python
new["comment_length"].fillna(0, inplace = True) 
new.isna().sum().plot(kind='bar')
```

>+ **Replacing missing values in days_since_last_review column with maximum number of that column**

```python
max_value=new['days_since_last_review'].max()
new['days_since_last_review'].fillna(value=max_value, inplace=True)
new
```

```python
new.isna().sum().plot(kind='bar')
```

>+ **Handling empty cells of #of bed rooms column**

```python
# First method: Replace empty values in column bedrooms with the mean of values in the same column
""""
#Finding the mean of the column bedrroms
mean_value=new['bedrooms'].mean()

#Replace empty values in column bedrooms with the mean of values in the same column
new['bedrooms'].fillna(value=mean_value, inplace=True)
new.isna().sum().plot(kind='bar')
""""
```

> Finding out the number of ammenities per listing id

```python
amm_new=new.groupby('listing_id').amenities.max()
amm_new=amm_new.apply(lambda x: len(x.split(',')))
amm_new
```

```python
amm_count=pd.DataFrame(amm_new)
amm_count
```

```python
df_new1= pd.merge(new,amm_count, on='listing_id',how='left')

df_new1.rename(columns = {'amenities_y':'num_of_amenities'}, inplace = True)
df_new1
```

> Checking the details of columns w.r.t empty cells of bedroom

```python
missing_value=df_new1[df_new1.bedrooms.isna()]

#Checing missing values w.r.t room_type
missing_value.groupby('room_type').count()['name'].plot(kind='bar')
```

```python
missing_index= df_new1[pd.isna(df_new1.bedrooms)].index
missing_index
```

```python
missing_predicted= df_new1[df_new1['bedrooms'].isna()]['accomodates'].map(acc_bed)
missing_predicted
```

```python
#Checing missing values w.r.t bedrooms
empty_value=df_new1[df_new1.beds.isna()]
empty_value.groupby('bedrooms').count()['name'].plot(kind='bar')
```

```python
#Filling those missing values

for i,j in enumerate(empty_index):
    df_new1.at[j,'beds']=empty_predicted.iloc[i]

df_new1
    
```
